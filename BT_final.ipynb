```python
#print the sources code of the current url 
soup
```

```python
#find the list of items with a specif class name
list_items = soup.find_all('div',{'class':'jss182'})
```

```python
#len of the list with the class name
len(list_items)
```

```python
list_items[0].prettify()
```

```python
#Product name 
list_items[0].find("div", {"id": "product-name"}).get_text()
```

```python
# Product price 
list_items[0].find("span", {"id": "product-price"}).get_text().replace('Î”','')
```

```python
#contract length
list_items[0].find("div", {"id": "contract-length"}).get_text()
```

```python
#upfromt cost
list_items[0].find("span", {"id": "upfront-cost"}).get_text().replace('\xa0','')
```

```python
#list comprehension python
#itreate through the list of items in list_items
product_title = [result.find("div", {"id": "product-name"}).get_text() for result in list_items]
product_price = [result.find("span", {"id": "product-price"}).get_text().replace('Î”','') for result in list_items]
contract_length = [result.find("div", {"id": "contract-length"}).get_text() for result in list_items]
upfront_cost = [result.find("span", {"id": "upfront-cost"}).get_text().replace('\xa0','') for result in list_items]
Downlading_speed = [result.find("span", {"class": "jss454"}).get_text() for result in list_items]
```

```python
#Creating a data frame with the data extracted
bt_deal_df = pd.DataFrame({'product_title': product_title, 'product_price': product_price,'contract_length': contract_length,'upfront_cost': upfront_cost, 'Downlading_speed': Downlading_speed})
```
